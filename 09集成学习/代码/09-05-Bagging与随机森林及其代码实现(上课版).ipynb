# bagging

```python
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

#### bootstrap = True 为bagging，bootstrap=False为pasting
#### max_samples设置为整数表示的就是采样的样本数，设置为浮点数表示的是max_samples*x.shape[0]

```python
bag_clf = BaggingClassifier(
    SVC(),
    n_estimators=500,
    bootstrap=True,
    max_samples=1.0,
    oob_score=True
)
```

```python
bag_clf.fit(X,y)
# y_hat = bag_clf.predict(X)
# print(bag_clf.__class__.__name__,'=',accuracy_score(y,y_hat))
print(bag_clf.oob_score_)
```

# 如果基分类器是决策树，那么就是随机森林

```python
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(),
    n_estimators=500,
    bootstrap=True,
    max_samples=1.0
)
bag_clf.fit(X,y)
y_hat = bag_clf.predict(X)
print(bag_clf.__class__.__name__,'=',accuracy_score(y,y_hat))
```

# Sklearn也提供了直接实现随机森林的API

```python
rnd_clf = RandomForestClassifier(n_estimators=500)
rnd_clf.fit(X, y)
print(rnd_clf.__class__.__name__, '=', accuracy_score(y, y_hat))
```
